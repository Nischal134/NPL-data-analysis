```python
import pandas as pd
```

```python
input_path = '/Users/nischal/Desktop/NPL/Teamwise polishing/Biratnagar/Biratnagar.csv'
df = pd.read_csv(input_path)

#importing data and initialization in dataframe.
```

```python
df['did_bat'] = df['balls'].notna().map({True: 'T', False: 'F'})
df['did_bowl'] = df['overs'].notna().map({True: 'T', False: 'F'})
''' Created new column did_bat and did_bowl, respective columns looks for null values in balls and overs column and 
 returns boolean values with T and F'''
```

```python
bat_cols = ['runs', 'balls', 'fours', 'sixes', 'sr']
bowl_cols = ['overs', 'maidens', 'runs_conceded', 'wickets', 'economy', 'wides', 'no_balls']
#Creating lists of column names. Now they represents entire bundle of columns instead of doing one column at a time.
```

```python
df[bat_cols] = df[bat_cols].fillna(0)
df[bowl_cols] = df[bowl_cols].fillna(0)
#Replaces all null values with 0. (Used 0 because cricket stats has numerical values, so can't use mean or min/max values)
```

```python
mask_valid_sr = (df['did_bat'] == 'T') & (df['balls'] > 0)
df.loc[mask_valid_sr, 'sr'] = (df.loc[mask_valid_sr, 'runs'] / df.loc[mask_valid_sr, 'balls'] * 100).round(2)
'''Created filter named mask_valid_sr to fetch records of players who actually batted
Also created SR column with formula runs/balls with float value upto 2'''
```

```python
mask_valid_economy = (df['did_bowl'] == 'T') & (df['overs'] > 0)
df.loc[mask_valid_economy, 'economy'] = (df.loc[mask_valid_economy, 'runs_conceded'] / df.loc[mask_valid_economy, 'overs']).round(2)
```

```python
output_path = '/Users/nischal/Desktop/NPL/Teamwise polishing/Biratnagar/biratnagar_cleaned.csv'
df.to_csv(output_path, index=False)
#Extraction of tables.
```
